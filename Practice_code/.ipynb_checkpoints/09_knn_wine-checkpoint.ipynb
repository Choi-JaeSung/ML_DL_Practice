```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import KFold, cross_val_score, train_test_split

from tqdm import tqdm
```

```python
red = pd.read_csv('./datasets/winequality-red.csv', sep=';') 
# red = pd.read_csv('./datasets/winequality-red.csv', delimiter=';') 
```

```python
white = pd.read_csv('./datasets/winequality-white.csv', sep=';')
```

```python
red
```

```python
white
```

```python
red['target'] = 0
white['target'] = 1
```

```python
white.duplicated().sum()
```

```python
white.drop_duplicates(keep="first", inplace=True)
white
```

```python
red.drop_duplicates(keep="first", inplace=True)
red
```

```python
wine = pd.concat([red, white], axis = 0)
```

```python
# wine = wine.drop(["fixed acidity", "volatile acidity", "chlorides", "density", "pH", "sulphates"], axis=1)
```

```python
wine
```

```python
X = wine.iloc[:, :-1]
y = wine.iloc[:, -1]
```

```python
# Train, Test 분리

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2022, stratify=y, shuffle=True)
```

```python
#표준화
scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
```

```python
y_train = y_train.to_numpy()
```

```python
hist = []

for k in tqdm(range(3, 50)):
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X_train_scaler, y_train, cv=5, scoring='accuracy')
    
    hist.append(scores.mean())
    
np.array(hist).mean()
```
