### Plot 여러개 표시 1번째: subplot(row, col, num)

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(xarr)
plt.subplot(1, 2, 2)
plt.plot(xarr)
plt.show()
```

### Plot 여러개 표시 2번째: fig.add_subplot(row, col, num)

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(xarr)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(xarr)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(xarr)
ax2.plot(yarr)
plt.show()
```

### Plot 여러개 표시 3번째: subplots(row, col)

```python
fig, axes = plt.subplots(2,2)
```

```python
axes
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(2,2)
axes[0][0].plot(xarr)
axes[1, 1].plot(yarr)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

axes = axes.flatten()
np.where(axes == axes[0])

axe_num = 1
for axe in axes:
    if axe_num % 2 == 1:
        axe.plot(xarr)
    else:
        axe.plot(yarr)
    axe_num += 1
    
# plt.show()
```

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(8, 6))

plt.title("This is title")
plt.xlabel("X label")
plt.ylabel("y label")

plt.plot(xarr, label="korea", color='r', linestyle='--', marker="o")
plt.plot(yarr, label="english", color='g', linestyle='-.', marker="^")
plt.plot(zarr, label="japanese", color='b', linestyle=":", marker="*")

plt.legend()
plt.show()
```

## Scatter plot

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.scatter(xarr, yarr, color="g", s=200)
plt.scatter(xarr, zarr, color='b', marker="^", s=100)
plt.show()
```

## Bar plot

```python
year = ["A", "B", "C"]
points = [100, 200, 300]
```

```python
plt.bar(year, points)
plt.show()
```

```python
plt.barh(year, points, color=["r", "g", "b"])
plt.show()
```

## Box plot

```python
x = np.random.normal(0, 2.0, 100)
y = np.random.normal(-3, 1.5, 500)
z = np.random.normal(1.2, 1.5, 1000)
```

```python
plt.boxplot([x, y, z])
plt.show()
```
