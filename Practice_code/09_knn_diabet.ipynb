```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import KFold, cross_val_score, train_test_split

from tqdm import tqdm
```

```python
dia = pd.read_csv('./datasets/diabetes.csv')
```

```python
dia.corr()
```

```python
dia = dia.drop(["BloodPressure", "SkinThickness"], axis=1)
```

```python
X = dia.iloc[:, :-1]
y = dia.iloc[:, -1]
```

```python
# Train, Test 분리

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2022, stratify=y, shuffle=True)
```

```python
#표준화
scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
```

minmax = MinMaxScaler()

```python
y_train = y_train.to_numpy()
```

```python
hist = []

for k in tqdm(range(3, 50)):
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X_train_scaler, y_train, cv=5, scoring='accuracy')
    
    hist.append(scores.mean())
    
np.array(hist).mean()
```

```python
print("최대 k값 : ",np.argmax(hist) + 3)
```
