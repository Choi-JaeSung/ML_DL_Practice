```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("./Dataset/titanic.csv")
```

## Histogram

```python
sns.histplot(data=titanic, x='Age', bins=10)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='Age', bins=10, hue= 'Sex')
plt.show()
```

## displot
* figure-level plot

```python
sns.displot(data=titanic, x='Age')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', kde=True)
plt.show()
```

```python
sns.displot(data=titanic, x='Age', kind='kde')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', col='Survived')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', col='Survived', hue='Sex')
plt.show()
```

## bar plot

```python
plt.bar(titanic['Pclass'].value_counts().index, titanic['Pclass'].value_counts().values)
plt.show()
```

```python
sns.countplot(data=titanic, x='Pclass')
plt.show()
```

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', ci=False)
plt.show()
```

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', ci=False, hue='Sex')
plt.show()
```

## Point plot

```python
sns.pointplot(data=titanic, x='Pclass', y='Fare', hue='Sex')
plt.plot()
```

```python
sns.boxplot(data=titanic, x='Pclass', y='Age', hue='Sex')
plt.plot()
```

## violin plot

```python
sns.violinplot(data=titanic, x='Pclass', y='Age', hue='Sex', split=True)
plt.plot()
```

## count plot

```python
sns.countplot(data=titanic, x='Embarked')
plt.plot()
```

## heatmap

```python
data = np.random.rand(4, 5)
data.shape
```

```python
sns.heatmap(data=data, annot=True)
plt.plot()
```

```python
sns.heatmap(data=data, annot=True, cmap='YlGnBu')
plt.plot()
```

## scatter plot

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.plot()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.plot()
```

## reg plot (regression)

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.plot()
```
